```python
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV

def MLpipe_KFold_Accuracy(X,y,preprocessor,algorithm,params):
    best_test_scores=[]
    best_models=[]
    
    for rs in range(10):
        X_other,X_test,y_other,y_test = train_test_split(X,y,train_size=.8,random_state=rs)
        kf=KFold(n_splits=4,shuffle=True,random_state=rs)
        pipeline=make_pipeline(preprocessor,algorithm)
        grid = GridSearchCV(pipeline, param_grid=params,scoring = 'accuracy',cv=kf)
        grid.fit(X_other, y_other)
        best_param =grid.best_params_
        best_models.append(grid.best_estimator_)
        y_test_pred = grid.predict(X_test)
        best_test_scores.append(accuracy_score(y_test,y_test_pred))
        
    return (best_models, best_test_scores)
```

```python
from sklearn.impute import SimpleImputer
from sklearn.pipeline import make_pipeline

def reduced_feature_pipeline(X, y, cat_ftrs, preprocessor, model, params):
    all_best_models=[]
    X_tests=[]
    y_tests=[]
    y_test_preds=[]
    test_accuracies=[]
    
    X[cat_ftrs] = X[cat_ftrs].to_string()
    imputer = ColumnTransformer(transformers=[
            ('onehot', SimpleImputer(strategy='constant',fill_value='Unknown'), cat_ftrs)],
                                       remainder='passthrough',verbose_feature_names_out=False)
    X_prep=pd.DataFrame(imputer.fit_transform(X), columns=X.columns)

    for rs in range(5):
        best_models=[]
        kf = KFold(n_splits=5,shuffle=True,random_state=rs)

        X_test,X_other,y_test,y_other=train_test_split(X_prep,y,train_size=.2,random_state=rs,shuffle=True)
        X_tests.append(X_test)

        mask = X_test.isnull()
        unique_rows = np.array(np.unique(mask, axis=0))
        all_y_test_pred = pd.DataFrame()
        
        for i in range(len(unique_rows)):
            X_test_subset = pd.DataFrame()
            y_test_subset = pd.Series(dtype=int)
            for j in range(len(mask)):
                row_mask=np.array(mask.iloc[j])
                if np.array_equal(row_mask,unique_rows[i]):
                    X_test_subset = X_test_subset.append(X_test.iloc[j])
                    y_test_subset = y_test_subset.append(y_test.iloc[[j]])
                    
            X_test_subset = X_test_subset[X_test.columns[~unique_rows[i]]]
            
            X_other_subset = pd.DataFrame(X_other.columns[~unique_rows[i]])
            X_other_subset = X_other_subset.dropna()
            y_other_subset = y_other.iloc[X_other_subset.index]
            
            pipeline=make_pipeline(preprocessor,model)
            grid = GridSearchCV(pipeline, param_grid=params,scoring = 'accuracy',cv=kf)
            grid.fit(X_other_subset, y_other_subset)
            best_params =grid.best_params_
            best_models.append(grid.best_estimator_)
            y_test_pred = grid.predict(X_test_subset)
            best_test_scores.append(accuracy_score(y_test,y_test_pred))
            all_y_test_pred = all_y_test_pred.append(y_test_subset_pred)
        
        all_best_models.append(best_models)
        all_y_test_pred = all_y_test_pred.sort_index()
        y_test = y_test.sort_index()
        y_tests.append(y_test)
        y_test_preds.append( all_y_test_pred)
        test_accuracies.append(accuracy_score(y_test,all_y_test_pred))
        
    return (all_best_models,test_accuracies,X_tests,y_tests,y_test_preds)
```
